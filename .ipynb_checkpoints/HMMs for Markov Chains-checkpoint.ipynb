In this notebook I have combined several resources I found about the application of Hidden Markov Models (HMMs) for making inference in Markov Chains. Focus is on the recovering of the transition probability matrix as well as model parameters in cases of full and partial observability, autoregressive and heteroskedastic process, with one or multiple sequences available. Finally the marginalized HMM is presented. We will exploit the PyMC3 library for a Bayesian estimation of HMMs.

The notebook hence focuses on the implementation of HMMs over a large range of different cases rather than covering the theory behind these tools. This work is mainly adapted from the great Eric Ma's [blogpost](https://ericmjl.github.io/essays-on-data-science/machine-learning/markov-models/), which is highly suggested for a practical understanding of Markov Chains and HMMs. 

## Data Generating Process

This section shows the data generating process of HMMs.

A Markov Chain is described by a number of possible states, an initial probability distribution and the transition probability matrix, from which is possible to compute the equilibrium distribution.

```python
n_states = 3

p_init = np.array([1/3., 1/3., 1/3.])
p_transition = np.array(
    [[0.90, 0.09, 0.01],
     [0.01, 0.90, 0.09],
     [0.01, 0.05, 0.94]]
)
p_transition
```

```python
assert p_transition[0, :].sum() == 1
assert p_transition[1, :].sum() == 1
assert p_transition[2, :].sum() == 1
```

```python
def equilibrium_distribution(p_transition):
    n_states = p_transition.shape[0]
    A = np.append(
        arr=p_transition.T - np.eye(n_states),
        values=np.ones(n_states).reshape(1, -1),
        axis=0
    )
    pinv = np.linalg.pinv(A)
    return pinv.T[-1]


print(equilibrium_distribution(p_transition))
```

```python
def markov_sequence(p_init: np.array, p_transition: np.array, sequence_length: int) -> List[int]:
    """
    Generate a Markov sequence based on p_init and p_transition.
    """
    if p_init is None:
        p_init = equilibrium_distribution(p_transition)
    initial_state = list(multinomial.rvs(1, p_init)).index(1)

    states = [initial_state]
    for _ in range(sequence_length - 1):
        p_tr = p_transition[states[-1]]
        new_state = list(multinomial.rvs(1, p_tr)).index(1)
        states.append(new_state)
    return states
```

```python
lenght_seq = 1000

states = markov_sequence(p_init, p_transition, sequence_length=lenght_seq)
fig, ax = plt.subplots(figsize=(12, 4))
plt.plot(states)
plt.xlabel("time step")
plt.ylabel("state")
plt.yticks([0, 1, 2])
sns.despine()
```

In practice, the states of Markov Chains are often not observable (hidden/latent variables), but they generate the observations (or emissions) according to a certain data generating processs. Here, we consider the case where the discrete states generate continuous observations. Each state is hence associated with a different Gaussian distribution. In addition, the current observation not only depends on the hidden state but also on the previous observation (autoregressive component with parameter $k$). Since the variance parameter is not constant among the different distributions, the process is also heteroskedastic. We consider this case as it is a very general one and other cases can simply be derived by hyperparameter tuning. For example, non-autoregressive processes are recovered by setting $k=0$ or homosckedasticity is obtained with $\sigma_i = \sigma \hspace{0.2cm} \forall i$. Finally, emissions are not in general restricted to come from Normal nor continuous distributions.

```python
def ar_gaussian_heteroskedastic_emissions(states: List[int], k: float, mus: List[float], sigmas: List[float]) -> List[float]:
    emissions = []
    prev_loc = 0
    for state in states:
        e = norm.rvs(loc=k * prev_loc + mus[state], scale=sigmas[state])
        emissions.append(e)
        prev_loc = e
    return emissions


def plot_emissions(states, emissions):
    fig, axes = plt.subplots(figsize=(16, 8), nrows=2, ncols=1, sharex=True)
    axes[0].plot(states)
    axes[0].set_title("States")
    axes[1].plot(emissions)
    axes[1].set_title("Emissions")
    sns.despine();
```

```python
k = 0.5
mus = [1, 0.5, -1]
sigmas = [1, 0.5, 0.2]

ar_het_ems = ar_gaussian_heteroskedastic_emissions(states, k, mus, sigmas)
plot_emissions(states, ar_het_ems)
```

## HMM 

The HMM is implemented below. Following Eric Ma's essay, we assume the initial state to be distributed according to the equilibrium probabilities.

```python
class HMMStates(pm.Categorical):
    def __init__(self, p_transition, p_equilibrium, n_states, *args, **kwargs):
        super(pm.Categorical, self).__init__(*args, **kwargs)
        self.p_transition = p_transition
        self.p_equilibrium = p_equilibrium
        self.k = n_states
        self.mode = tt.cast(0,dtype='int64')

    def logp(self, x):
        p_eq = self.p_equilibrium
        p_tr = self.p_transition[x[:-1]]

        # the logp of the initial state evaluated against the equilibrium probabilities
        initial_state_logp = pm.Categorical.dist(p_eq).logp(x[0])

        # the logp of the rest of the states.
        x_i = x[1:]
        ou_like = pm.Categorical.dist(p_tr).logp(x_i)
        transition_logp = tt.sum(ou_like)
        return initial_state_logp + transition_logp
```

```python
def solve_equilibrium(n_states, p_transition):
    A = tt.dmatrix('A')
    A = tt.eye(n_states) - p_transition + tt.ones(shape=(n_states, n_states))
    p_equilibrium = pm.Deterministic("p_equilibrium", sla.solve(A.T, tt.ones(shape=(n_states))))
    return p_equilibrium
```

Before going into the "hidden" case, we show how the model is applied when the states are fully observable. The objective is to correctly recover the transition probability matrix, which is given a Dirichlet prior (generalization of the Beta distribution).

```python
with pm.Model() as model:
    transition_mat = pm.Dirichlet(
        "p_transition",
        a=tt.ones((n_states, n_states)), 
        shape=(n_states, n_states))

    # Solve for the equilibrium state
    p_equilibrium = solve_equilibrium(n_states, p_transition)

    obs_states = HMMStates(
        "states",
        p_transition=transition_mat,
        p_equilibrium=p_equilibrium,
        n_states=n_states,
        observed=np.array(states).astype("float")
    )
    
    trace = pm.sample(2000)
```

```python
az.plot_trace(trace,  var_names=["p_transition"]);
```

```python
az.plot_posterior(trace, var_names=["p_transition"]);
```

The MCMC sampling worked smoothly and the probability entries of the transition matrix are recovered successfully.

## ARHMM Gaussian Emissions

We now consider the case of the Gaussian emissions generated by the hidden states. Below the observation generating process is implemented. 

```python
class ARHMMGaussianEmissions(pm.Continuous):
    def __init__(self, states, k, mu, sigma, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.states = states
        self.sigma = sigma 
        self.mu = mu
        self.k = k  

    def logp(self, x):
        """
        x: observations
        """
        states = self.states
        mu = self.mu[states]
        sigma = self.sigma[states]
        k = self.k

        ar_mean = k * x[:-1]
        ar_like = tt.sum(pm.Normal.dist(mu=ar_mean + mu[1:], sigma=sigma[1:]).logp(x[1:]))

        boundary_like = pm.Normal.dist(mu=mu[0], sigma=sigma[0]).logp(x[0])
        return ar_like + boundary_like
```

When estimating the model, rather than assigning the usual prior $\mu \sim \mathcal{N}(0,1)$, we give different priors to $\mu$ in order to avoid intra-chain omogeneity (non-identifiability). The model is hence summarized as follows:

$$\mathcal{S} = \{0, 1, 2\}
\\
\\
s_t \in S
\\
\\
s_t | s_{t-1} \sim Dirichlet(\alpha)
\\
\\
\mu_{s_t} \sim \mathcal{N}(\bar{\mu}_i, 1)
\\
\\
\bar{\mu}_i = \{0.5, 0, -0.5\}, i \in S
\\
\\
\sigma_{s_t} \sim Exp(2) \hspace{0.2cm} \forall i \in S
\\
\\
k \sim Beta(2,2)
\\
\\
y_t | s_t \sim \mathcal{N}(ky_{t-1} + \mu_{s_t}, \sigma_{s_t})    $$

```python
with pm.Model() as model:
    # Priors for transition matrix
    transition_mat = pm.Dirichlet("p_transition", a=tt.ones((n_states, n_states)), shape=(n_states, n_states))

    # Solve for the equilibrium state
    p_equilibrium = solve_equilibrium(n_states, transition_mat)

    # HMM state
    hmm_states = HMMStates(
        "hmm_states",
        p_transition=transition_mat,
        p_equilibrium=p_equilibrium,
        n_states=n_states,
        shape=(len(ar_het_ems),)
    )

    # Prior for mu, sigma and k
    μ = pm.Normal("mu", mu=[0.5, 0., -0.5], sigma=1, shape=(n_states,)) # different priors for the 3 Normal dist
    σ = pm.Exponential("sigma", lam=2, shape=(n_states,))
    k_p = pm.Beta("k", alpha=2, beta=2) 

    # Observed emission likelihood
    obs = ARHMMGaussianEmissions(
        "emission",
        states=hmm_states,
        mu=μ,
        sigma=σ,
        k=k_p,
        observed=ar_het_ems
    )
    
    nuts_step = pm.NUTS([model.k, model.sigma, model.mu, model.p_transition], target_accept=0.9)
    trace = pm.sample(2000, step=[nuts_step])
```

```python
az.plot_trace(trace,  var_names=["p_transition", "mu", "sigma", "k"]);
```

```python
az.plot_posterior(trace, var_names=["p_transition"]);
```

```python
az.plot_posterior(trace, var_names=["mu", "sigma", "k"]);
```

```python
az.plot_forest(trace, var_names=["mu", "sigma"]);
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
plt.plot(np.round(trace["hmm_states"].mean(axis=0)), label="estimates")
plt.plot(np.array(states), label="true")
plt.legend();
```

Despite the partial observability, the HMM correctly inferred the transition probability matrix, model parameters and the hidden states.

## Multiple Sequence

When dealing with real data, it is likely to have multiple (short) sequences rather than one longer. In this section we show which changes are needed to handle multiple sequences. This section is adapted from this [discussion](https://discourse.pymc.io/t/hidden-markov-model-estimating-transition-and-emission-cpds-from-multiple-sequences-not-working/6308/11) on PyMC3 forum.

```python
def generate_multiple_seq(lenght_seq, number_seq, p_transition, p_init, k, mus, sigmas, plot=True): 
    #Generating a Markov Sequence
    states_all_seq = np.full((number_seq, lenght_seq), np.nan)
    for i in range(number_seq):
        states = markov_sequence(p_init, p_transition, sequence_length=lenght_seq)
        states_all_seq[i, :] = states
    states_all_seq = states_all_seq.astype(int)
    # Generate Emissions: states and observables
    emissions_all_seq = np.full((number_seq, lenght_seq), np.nan)
    for i in range(number_seq):
        emissions = ar_gaussian_heteroskedastic_emissions(states_all_seq[i, :], k, mus, sigmas)
        emissions_all_seq[i, :] = emissions
        if plot:
            plot_emissions(states_all_seq[i, :], emissions_all_seq[i, :])
    
    return states_all_seq, emissions_all_seq
```

```python
lenght_seq = 1000
number_seq = 2
k = 0.5
mus = [1, 0.5, -1]
sigmas = [1, 0.5, 0.2]

states_all_seq, emissions_all_seq = generate_multiple_seq(lenght_seq, number_seq, p_transition, p_init, k, mus, sigmas)
```

### Full Observability

First, the case of fully observable states is presented.

```python
with pm.Model() as model:
    transition_mat = pm.Dirichlet(
        "p_transition",
        a=tt.ones((n_states, n_states)), 
        shape=(n_states, n_states))
    
    # Solve for the equilibrium state
    p_equilibrium = solve_equilibrium(n_states, transition_mat)
    
    for i in range(number_seq):
        obs_states = HMMStates(
            "states_" + str(i),
            p_transition=transition_mat,
            p_equilibrium=p_equilibrium,
            n_states=n_states,
            observed=np.array(states_all_seq[i, :]).astype("float")
        )
    
    trace = pm.sample(2000, tune=1000)
```

```python
az.plot_trace(trace,  var_names=["p_transition"]);
```

```python
az.plot_posterior(trace,  var_names=["p_transition"]);
```

The inferred probabilities seem even closer to the true values compared to the one sequence case.

### Partial Observability

We now present the case with hidden states and Gaussian emissions.

```python
with pm.Model() as model:
    # Priors for transition matrix
    transition_mat = pm.Dirichlet("p_transition", a=tt.ones((n_states, n_states)), shape=(n_states, n_states))

    # Solve for the equilibrium state
    p_equilibrium = solve_equilibrium(n_states, transition_mat)
    
    # Prior for mu, sigma and k
    μ = pm.Normal("mu", mu=[0.5, 0., -0.5], sigma=1, shape=(n_states,)) 
    σ = pm.Exponential("sigma", lam=2, shape=(n_states,))
    k_p = pm.Beta("k", alpha=2, beta=2) 
    
    for i in range(number_seq):
        # HMM state
        hmm_states_i = HMMStates(
            "hmm_states_" + str(i),
            p_transition=transition_mat,
            p_equilibrium=p_equilibrium,
            n_states=n_states,
            shape=(lenght_seq,)
        )


        # Observed emission likelihood
        obs_i = ARHMMGaussianEmissions(
            "emission_i" + str(i),
            states=hmm_states_i,
            mu=μ,
            sigma=σ,
            k=k_p,
            observed=np.array(emissions_all_seq[i, :]).astype("float")
        )
        
    nuts_step = pm.NUTS([model.k, model.sigma, model.mu, model.p_transition], target_accept=0.9) 
    trace = pm.sample(2000, step=[nuts_step])
```

```python
az.plot_trace(trace,  var_names=["p_transition", "mu", "sigma", "k"]);
```

```python
az.plot_posterior(trace,  var_names=["p_transition"]);
```

```python
az.plot_posterior(trace,  var_names=["mu", "sigma", "k"]);
```

```python
fig, ax = plt.subplots(2, 1, figsize=(12, 4*number_seq))
for i in range(number_seq):
    ax[i].plot(np.round(trace["hmm_states_"+str(i)].mean(axis=0)), label="estimates")
    ax[i].plot(np.array(states_all_seq[i, :]), label="true")
    ax[i].legend()
```

Once again, all parameters are correctly estimated and the HMM successfully handled multiple sequences.

## Marginalized HMM

HMMStates is a discrete (Categorical) distribution and it is generally not recommended working with discrete latent variables because they cannot be sampled efficiently using NUTS. Hence, here we present the marginalized HMM based on this [notebook](https://github.com/junpenglao/Planet_Sakaar_Data_Science/blob/master/PyMC3QnA/discourse_2230.ipynb) where the discrete latent variables are marginalized out. This can be an excellent variant if one is interested in recovering the transition matrix and all other parameters except hidden states. 

### One sequence

First we apply the marginalized HMM for one sequence. In order to simplify the code, we deal with the non-autoregressive case.

```python
k = 0.
mus = [1, 0.5, -1]
sigmas = [1, 0.5, 0.2]

het_ems = ar_gaussian_heteroskedastic_emissions(states, k, mus, sigmas)
plot_emissions(states, ar_het_ems)
```

```python
observed_label = theano.shared(np.array(het_ems))

def onestep(obs, gamma_, theta, mu, sigma):
    """
    One step of the forward algorithm.
    """
    i = obs
    alpha = gamma_ + tt.log(theta) + tt.tile(
        pm.Normal.dist(mu=mu,sigma=sigma).logp(i), (theta.shape[0], 1))
    return pm.math.logsumexp(alpha, axis=0).T

with pm.Model() as model:
    transition_mat = pm.Dirichlet("p_transition", a=tt.ones((n_states, n_states)), shape=(n_states, n_states))
    
    # Prior for mu and sigma
    μ = pm.Normal("mu", mu=[0.5, 0., -0.5], sigma=1, shape=(n_states,))
    σ = pm.Exponential("sigma", lam=2, shape=(n_states,))
     
    gamma=pm.Normal.dist(mu=μ,sigma=σ).logp(het_ems[0])
    gamma = tt.tile(gamma, (1, 1)).T

    result, updates = theano.scan(fn=onestep,
                                  outputs_info=gamma,
                                  sequences=observed_label,
                                  non_sequences=[transition_mat, μ, σ],
                                  n_steps=lenght_seq-1)
    
    obs_logp = pm.Potential('obs_logp', pm.math.logsumexp(result[-1]))
    
    trace = pm.sample(2000, tune=1000, target_accept=0.9) 
```

```python
az.plot_trace(trace);
```

```python
az.plot_posterior(trace, var_names=["p_transition"]);
```

```python
az.plot_forest(trace, var_names=["mu", "sigma"]);
```

The parameters are still correctly estimated, but the inference took longer and the distributions do not appear to have reached convergence. The chains are also affected by non-identifiability.

### Multiple sequence

In this last case, we apply the marginalized HMM to multiple sequences without the autoregressive component.

```python
states_all_seq, emissions_all_seq = generate_multiple_seq(lenght_seq, number_seq, p_transition, p_init, k, mus, sigmas)
```

```python
observed_all_seq = theano.shared(np.array(emissions_all_seq))

def onestep(obs, gamma_, theta, mu,sigma):
    """
    One step of the forward algorithm.
    """
    i = obs
    alpha = gamma_ + tt.log(theta) + tt.tile(
        pm.Normal.dist(mu=mu,sigma=sigma).logp(i), (theta.shape[0], 1))
    return pm.math.logsumexp(alpha, axis=0).T


with pm.Model() as model:
    transition_mat = pm.Dirichlet("p_transition", a=tt.ones((n_states, n_states)), shape=(n_states, n_states))

    # Prior for mu and sigma
    μ = pm.Normal("mu", mu=[0.5, 0., -0.5], sigma=1, shape=(n_states,))
    σ = pm.Exponential("sigma", lam=2, shape=(n_states,))

    aux = 0
    for i in range(number_seq):
        gamma=pm.Normal.dist(mu=μ,sigma=σ).logp(emissions_all_seq[i, 0])
        gamma = tt.tile(gamma, (1, 1)).T
        result, updates = theano.scan(fn=onestep,
                                      outputs_info=gamma,
                                      sequences=observed_all_seq[i, :],
                                      non_sequences=[transition_mat, μ, σ],
                                      n_steps=lenght_seq-1)
        aux += pm.math.logsumexp(result[-1])
    
    obs_logp = pm.Potential('obs_logp', aux)
    trace = pm.sample(2000, tune=1000, target_accept=0.9) 
```

```python
az.plot_trace(trace, var_names=["p_transition", "mu", "sigma"]);
```

```python
az.plot_posterior(trace, var_names=["p_transition"]);
```

```python
az.plot_forest(trace, var_names=["mu", "sigma"]);
```

Also for the multiple sequence case, the marginalized HMM does not seem to work better than the non-marginalized HMM (it actually looks rather worse). I cannot exclude that I made some mistakes in the adaptation, but the inference was not very good in the reference notebook either.
